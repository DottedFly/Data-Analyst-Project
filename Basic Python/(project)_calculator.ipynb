```python
usage_count = 0

print('hi and welcome to simplycalt, enter your name please')

name = (input('please input your name'))
if name:
    print(f'hi {name}, nice to meet you')


while True :
    print('put your desire number')
    a = int(input('input your desire number'))
    print('put your second desire number')
    b = int(input('input your second desire number'))
    print('data stored')
    print(f'your datas are {a} and {b}')
    print('please select your desire operation (+|-|*|/) or want to stop')

    operation = input('''
    please select the operation
    ( + ) for addition
    ( - ) for substraction
    ( * ) for multiplication
    ( / ) for division
    or 
    type 'stop' to shutdown

    ''')

    if operation.lower() == 'stop' :
        print('shutting down calculator')
        break




    match operation:
        case '+' :
            print('addition')
            print(f'result : {a + b}' )  
            usage_count += 1
            continue

        case '-' :
            print('substraction')
            print(f'result : {a - b}' )
            usage_count += 1
            continue

        case '*' :
            print('multiplication')
            print(f'result : {a * b}' )
            usage_count += 1
            continue
        
        case '/' :
            print('division')
            if b == 0 :
                print('cant divide by zero, try again')
                continue

            if a == 0 :
                print('cant divide by zero, try again')
                continue

            else :
                print(f'result : {a / b}' )
                usage_count += 1
                continue

print('heres how many you use us {usage_count}')
print('come back later')
```

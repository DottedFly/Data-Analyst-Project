<h3> Reducing unnecessary columns </h3>

```python
data
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data_delete = [
    'Unnamed: 0', 'asin', 'product_availability', 'unit_price', 
    'unit_count', 'product_photo', 'product_url', 
    'is_amazon_choice', 'is_prime', 'climate_pledge_friendly', 
    'sales_volume', 'delivery', 'has_variations','currency'
]
```

reason for dropping columns:
- 'Unnamed: 0' and 'asin' unable to identify
- unit_price, unit_count, product_availability almost all values are NaN
- product_photo, product_url is useless for analysis
- is_amazon_choice, is_prime, climate_pledge_friendly, delivery are all unnecessary for analysis
- has_variations also not useful for analysis
- sales_volume is kinda complicated 
- 'currency' all values are 'USD'

```python
data = data.drop(columns= data_delete)
```

```python
data
```

<h3> changing data types </h3>

```python
data['Is_best_seller'].value_counts()
```

```python
# now we gonna change True n False to 1 and 0
data['Is_best_seller'] = data['Is_best_seller'].apply(lambda x: 1 if x == True else 0)
#print(data['Is_best_seller'].value_counts())

#alternative turn it into int, same result 
#data['Is_best_seller'].astype(int)
#print(data['Is_best_seller'].value_counts())
```

```python
data['Product_price'] = data['Product_price'].str.replace('$', '')
```

```python
# also dont forget to change int to actual float
data['Product_price'] = data['Product_price'].astype(float)
```

```python
# double check
type(data['Product_price'][0])
```

```python
# or we can just round it
data['Product_price'] = data['Product_price'].apply(lambda x: round(x))
```

```python
# always double check everthing
type(data['Product_star_rating'][0])
```

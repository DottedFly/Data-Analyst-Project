Based on what i see, some needs adjustment :
- Host identity verified (from object to boolean)
- Instant bookable (from object to boolean)
- Construction Year (from float to int)
- Price (from object to float)
- Service fee (from object to float)


```python
#data['Host identity verified'].value_counts()
data['Instant bookable'].value_counts()
```

```python
data.columns
```

```python
data['Host identity verified'] = data['Host identity verified'].apply(lambda x: 1 if x == 'verified' else 0)
```

```python
data['Instant bookable'] = data['Instant bookable'].apply(lambda x: 1 if x == True else 0)

# 'True' != True , lesson learned
```

```python
data['Construction year'] = data['Construction year'].astype(int)
```

```python
for lo in ['Price', 'Service fee']:
#    print(data[lo].replace('[\$,]', '', regex=True ).astype(float))
    data[lo] = data[lo].replace('[\$,]', '', regex=True).astype(float)
```
